```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
 class Board:
    
     def __init__(self):
         """Instantiate and show an 8x8 Chess Board"""
         self.grid = np.zeros((8, 8, 3))
         white = (1,1,1)
         for i in range(8):
            for j in range(8):
                if ((i+j)%2):
                    self.grid[i][j] = white
         plt.imshow(self.grid)
        

     def add_red(self, row=None, col=None):
         """method input: an int row and an int col, output: places red square at input row and col and updates dictionary for future reference"""
         self.grid[row][col] = (1,.2,0)
         self.red = {'row':row, 'col':col}
       

     def add_blue(self, row=None, col=None):
         """method input: an int row and an int col, output: board with blue square at input row and col and updates dictionary for future reference"""
         self.grid[row][col] = (0,1,1)
         self.blue = {'row':row, 'col':col}
         

     def render(self):
         """method displays board with red and blue square at designated row/col if methods above are invoked prior"""
         plt.imshow(self.grid)
        

     def is_under_attack(self):
         """method output: boolean TRUE if blue and red squares are inline horizonally, vertically"""
         if self.red['row'] == self.blue['row'] or self.red['col'] == self.blue['col']:
            print ('True')
            return True

         """method output: boolean TRUE if blue and red squares are inline diagonally"""
         if abs(self.red['row']-self.blue['row']) == abs(self.red['col']-self.blue['col']):
             print('True')
             return True

         """method output: boolean FALSE if blue and red squares are not inline horizonally, vertically, or diagonally"""
         print('False')
         return False
        
 
     
```

```python
    """Render b-w board"""
    board = Board()
    
```

```python
"""Place red square"""

board = Board()
board.add_red(6,4)
board.render()
```

```python
"""Place blue square"""
board = Board()
board.add_red(6,3)
board.add_blue(7,0)
board.render()
```

```python
"""Test Queens in same column"""
board = Board()
board.add_red(6,2)
board.add_blue(0,2)
board.render()
assert board.is_under_attack()==True
```

```python
"""Test Queens in same column"""
board = Board()
board.add_red(0,4)
board.add_blue(4,4)
board.render()
assert board.is_under_attack()==True
```

```python
"""Test Queens in same row"""
board = Board()
board.add_red(7,6)
board.add_blue(7,4)
board.render()
assert board.is_under_attack()==True
```

```python
"""Test Queens in same row"""
board = Board()
board.add_red(5,3)
board.add_blue(5,7)
board.render()
assert board.is_under_attack()==True
```

```python
"""Test Queens in diagonal without needing abs()"""
board = Board()
board.add_red(5,3)
board.add_blue(3,1)
board.render()
assert board.is_under_attack()==True
```

```python
"""Test Queens in diagonal with abs() in effect"""
board = Board()
board.add_red(5,1)
board.add_blue(7,3)
board.render()
assert board.is_under_attack()==True
```

```python
"""Test Queens not in attack position"""
board = Board()
board.add_red(5,3)
board.add_blue(7,7)
board.render()
assert board.is_under_attack()==False
```
